```python
import pandas as pd
import numpy as np

#Criando um dicionário de Séries
d = {'Nomes':pd.Series(['Tom','Tom','Ricky','Vin','Steve','Smith','Jack',
   'Lee','David','Gasper','Betina','Andres']),
   'Idade':pd.Series([25,25,25,30,29,23,34,30,51,46]),
   'Qualidade':pd.Series([9,8,3.98,4.5,4.20,4.6,3.8,3.78,2.98,4.80,4.10,3.65]),
   'Projetos_Nota':pd.Series([8,7,3,4,3,4,3,3,2,4,45,3]),
    }

#Criando um DataFrame
df = pd.DataFrame(d)
print(df)
```

```python
df.Idade = df.Idade.fillna('-99') #Padronizar os valores vazios
```

```python
#data.dropna(axis=1, how=’all’) #dropa as colunas com todos os campos vazios

#data.dropna(axis=1, how=’any’) #dropa as colunas com pelo menos um valor vazio

#df.dropna() #dropa os registros que contém qualquer coluna vazia

df.dropna(how='all') #Dropar os registros que tem todos as colunas vazias
```
